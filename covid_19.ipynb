```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime as dt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('dataset\covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
# keeping only required column 
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
```

```python
df.head()
```

```python
# renaming column name 
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
df.tail()
```

```python
# last_date
last=df[df.date=='2020-07-01']
```

```python
last.head()
```

```python
max_confirmed_cases=last.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
top_states_confirmed=max_confirmed_cases[0:5]
```

```python
top_states_confirmed
```

```python
# making bar-plot for sates with top confirmed cases
sns.set(rc={'figure.figsize':(12,8)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
plt.title('Confirmed cases',color='blue')
plt.show()
```

```python
max_deaths_cases=last.sort_values(by='deaths',ascending=False)
max_deaths_cases
```

```python
top_states_deaths=max_deaths_cases[0:5]
top_states_deaths
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.barplot(x='state',y='deaths',data=top_states_deaths,hue='state')
plt.title('Top Deaths State',color='red')
plt.show()
```

```python
max_cured_cases=last.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
top_states_cured=max_cured_cases[0:5]
top_states_cured
```
